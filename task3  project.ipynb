```python
import pandas as pd
arr=pd.read_csv('foundational stage\\task3 resources\\day_wise.csv',encoding='latin1')
arr.head()
```

```python
arr.groupby("Date")["Confirmed"].mean().plot(kind='line', title='Day Wise Confirmed Cases')
```

```python
arr.groupby("Date")["Recovered"].sum().plot(kind='line')
```

insights
Global Case Insights
- Top confirmed cases:
- US (4.29M), Brazil (2.44M), India (1.48M), Russia (0.81M).
- Active cases hotspots:
- US (2.8M), Brazil (0.5M), India (0.49M).
- Rapid weekly growth:
- India (+324k, +28%), US (+455k, +12%), Brazil (+323k, +15%), Colombia (+53k, +26%).

Mortality Insights
- Highest death tolls:
- US (148k), Brazil (87k), Mexico (44k), UK (45k).
- Highest death rates (Deaths/100 cases):
- Yemen (28.6%), UK (15.2%), Italy (14.3%), Belgium (14.8%).
- Lowest death rates:
- Qatar (0.15%), Singapore (0.05%), Bahrain (0.36%).

Recovery Insights
- Highest recovery rates:
- Djibouti (98%), Brunei (97.9%), Qatar (97%).
- Lowest recovery rates:
- US (30.9%), UK (0.48%), Sweden (0%).

Regional Analysis
- Americas: US, Brazil, Mexico, Peru, Colombia dominate global case counts.
- Europe: UK, Spain, Italy, Russia  high deaths but mixed recovery reporting.
- Africa: South Africa (452k) is the epicenter; Ethiopia, Nigeria, Ghana rising fast.
- Eastern Mediterranean: Iran, Iraq, Saudi Arabia, Pakistan  significant case loads.
- South-East Asia: India (1.48M) dominates; Bangladesh, Indonesia also high.

Weekly Trends
- Explosive growth:
- Papua New Guinea (+226%), Gambia (+191%), Zimbabwe (+58%), Ethiopia (+42%).

Correlations
- Deaths vs. Recoveries: Countries with high death/recovery ratios (Mexico, UK, Yemen) show strained healthcare.
- Active vs. New cases: US, India, Brazil still in acceleration phase.

Key Takeaways
- Epicenters: US, Brazil, India drive global case growth.
- High recovery clusters: Qatar, Djibouti, Brunei.
- Regional hotspots: Americas and South-East Asia dominate global surge.
- Weekly surges: India, Colombia, Ethiopia, Zimbabwe show rapid acceleration.
